```python
# HeatMapAnalysis
plt.figure(figsize=(30,25))
sns.heatmap(data.drop("PerformanceRating",axis=1).corr(),cmap=plt.cm.Spectral_r,annot=True)
plt.title("HeatMapAnalysis")
plt.xlabel("columns_x",fontsize=10)
plt.ylabel("columns_y",fontsize=10)
plt.show()
```

This heatmap presents the pairwise correlation between all numerical features in the dataset, excluding the target variable PerformanceRating. Each cell in the matrix quantifies the strength and direction of the linear relationship between two features, ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation).
- Color Gradient: The Spectral_r colormap highlights strong correlations with vibrant contrast, making it easy to spot highly related features.
- Annotations: Each cell is labeled with its correlation coefficient for precise interpretation.
- Axes: columns_x and columns_y represent the feature names being compared.
This analysis is critical for:
- Identifying multicollinearity that may affect model performance
- Selecting or engineering features for predictive modeling
- Understanding how operational metrics (e.g., years of experience, salary hike) interact

```python
def correlation(dataset,threshold):
  col_corr=set()
  matrix=dataset.corr()
  for i in range(len(matrix.columns)):
    for j in range(i):
      if(matrix.iloc[i,j]>threshold):
        col_name=matrix.columns[i]
        col_corr.add(col_name)
      return col_corr
```

```python
features=correlation(data,0.7)
len(set(features))
```
