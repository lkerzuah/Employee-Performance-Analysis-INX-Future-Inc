# Project Summary

This data science project analyzes employee performance at INX Future Inc., with the goal of predicting performance ratings based on various employee attributes such as total work experience, gender, department, and current role.

**Project Goals & Insights**

- Department-wise performance analysis
- Identification of the top 3 factors influencing employee performance
- Development of a predictive model to assist in hiring decisions
- Actionable recommendations to improve employee performance based on data-driven insights


# Dataset Summary
- Source: Provided by IABAC™ 
- Platform: Jupyter Notebook using Python
- Size: 1,200 records × 28 features
- Feature Types:
    - 16 categorical features
    - 11 numerical features
    - 1 alphanumeric feature (Employee ID – excluded from modeling)
- Label: PerformanceRating (categorical)

# Machine Learning Approach
- Model Used: Random Forest Classifier
- Accuracy Achieved: ~93%
- Preprocessing:
    - One-hot encoding for categorical variables
    - Feature importance analysis to identify key predictors
- Key Insight: Numerical features showed stronger correlation with performance ratings
